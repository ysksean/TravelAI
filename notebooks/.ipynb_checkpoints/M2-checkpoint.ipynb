# Îç∞Ïù¥ÌÑ∞ Î°úÎìú

```python
import os
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, ElectraForSequenceClassification
from torch.optim import AdamW
from tqdm.notebook import tqdm  # Ï£ºÌîºÌÑ∞Ïö© ÏßÑÌñâÎ∞î

# ==========================================
# 1. ÏÑ§Ï†ï (Configuration)
# ==========================================
# ÌÖåÏä§Ìä∏Ìï† ÎïåÎäî 1~2Î≤àÎßå, Ïã§Ï†ú ÌïôÏäµÌï† ÎïåÎäî 10Î≤à Ïù¥ÏÉÅ ÏÑ§Ï†ïÌïòÏÑ∏Ïöî.
EPOCHS = 4              # [Î≥ÄÍ≤Ω] 2 -> 4 (Ïã¨Ìôî ÌïôÏäµÏùÑ ÏúÑÌï¥ Ï°∞Í∏à ÎäòÎ¶º)
LEARNING_RATE = 5e-5    
BATCH_SIZE = 16       
MAX_LEN = 128           
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ÌååÏùº Í≤ΩÎ°ú (ÌòÑÏû¨ Ï£ºÌîºÌÑ∞ ÎÖ∏Ìä∏Î∂Å ÏúÑÏπò Í∏∞Ï§Ä)
BASE_DIR = os.getcwd()
DATA_FILE = "../ERP ÌïÑÏöîÌïú Îç∞Ïù¥ÌÑ∞/6. ÎûúÎìúÏÇ¨ÌïúÌÖå Í≥†Í∞ù ÏöîÏ≤≠ ÎãµÎ≥Ä ÎÇ¥Ïö©_ÏßëÏóêÏÑúÏ∂îÍ∞Ä/3Ï§ëÎ∂ÑÎ•ò_ÏÉùÏÑ±.csv"
MODEL_SAVE_DIR = os.path.join(BASE_DIR, "models", "koelectra_sentiment")

# ÎùºÎ≤® ÎßµÌïë (0:Î∂àÍ∞Ä, 1:ÌôïÏ†ï, 2:Î≥¥Î•ò)
LABEL_MAP = {0: "Î∂àÍ∞Ä(Î∂ÄÏ†ï)", 1: "ÌôïÏ†ï(Í∏çÏ†ï)", 2: "Î≥¥Î•ò(Ï§ëÎ¶Ω)"}

print(f"üöÄ ÏÇ¨Ïö© Ïû•Ïπò: {DEVICE}")

# ==========================================
# 2. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è ÌôïÏù∏
# ==========================================
if not os.path.exists(DATA_FILE):
    print(f"‚ùå Ïò§Î•ò: '{DATA_FILE}' ÌååÏùºÏù¥ ÏóÜÏäµÎãàÎã§. ÏóÖÎ°úÎìúÌï¥Ï£ºÏÑ∏Ïöî.")
else:
    df = pd.read_csv(DATA_FILE)
    print(f"‚úÖ Îç∞Ïù¥ÌÑ∞ Î°úÎìú ÏÑ±Í≥µ! Ï¥ù {len(df)}Í∞ú")
    print(df.head()) # ÏÉÅÏúÑ 5Í∞ú Ï∂úÎ†•
```

# Î™®Îç∏ Ï§ÄÎπÑ

```python
# ==========================================
# 3. Îç∞Ïù¥ÌÑ∞ÏÖã ÌÅ¥ÎûòÏä§ Ï†ïÏùò
# ==========================================
class SentimentDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        row = self.data.iloc[index]
        text = str(row['text'])
        label = int(row['label'])

        # ÌÜ†ÌÅ¨ÎÇòÏù¥Ïßï (Î¨∏Ïû• -> Ïà´Ïûê)
        # padding='max_length'Î°ú Í∏∏Ïù¥Î•º 128Î°ú Í≥†Ï†ïÌïòÏó¨ ÏóêÎü¨ Î∞©ÏßÄ
        encoding = self.tokenizer.encode_plus(
            text,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# ==========================================
# 4. Î™®Îç∏ Î∞è ÌÜ†ÌÅ¨ÎÇòÏù¥Ï†Ä Î°úÎìú
# ==========================================
print("‚è≥ Î™®Îç∏ Î°úÎî© Ï§ë...")
model_name = "monologg/koelectra-base-v3-discriminator"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# num_labels=3 (Î∂àÍ∞Ä, ÌôïÏ†ï, Î≥¥Î•ò)
model = ElectraForSequenceClassification.from_pretrained(model_name, num_labels=3)
model.to(DEVICE)

# Îç∞Ïù¥ÌÑ∞ Î°úÎçî Ïó∞Í≤∞
dataset = SentimentDataset(df, tokenizer, MAX_LEN)
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

print("‚úÖ Î™®Îç∏ Ï§ÄÎπÑ ÏôÑÎ£å!")
```

# ÌïôÏäµ

```python
from tqdm import tqdm
# ==========================================
# 5. ÌïôÏäµ Î£®ÌîÑ (Training Loop)
# ==========================================
print(f"üî• ÌïôÏäµ ÏãúÏûë (Ï¥ù {EPOCHS}Ìöå)")
model.train()

for epoch in range(EPOCHS):
    total_loss = 0
    progress_bar = tqdm(loader, desc=f"Epoch {epoch+1}/{EPOCHS}")
    
    for batch in progress_bar:
        input_ids = batch['input_ids'].to(DEVICE)
        attention_mask = batch['attention_mask'].to(DEVICE)
        labels = batch['labels'].to(DEVICE)

        optimizer.zero_grad()
        
        # ÏàúÏ†ÑÌåå (Forward)
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        progress_bar.set_postfix({'loss': loss.item()})

    avg_loss = total_loss / len(loader)
    print(f"   Epoch {epoch+1} ÌèâÍ∑† Ïò§Ï∞®(Loss): {avg_loss:.4f}")

print("\nüéâ ÌïôÏäµ ÏôÑÎ£å!")

# ==========================================
# 6. Î™®Îç∏ Ï†ÄÏû•
# ==========================================
if not os.path.exists(MODEL_SAVE_DIR):
    os.makedirs(MODEL_SAVE_DIR)

model.save_pretrained(MODEL_SAVE_DIR)
tokenizer.save_pretrained(MODEL_SAVE_DIR)
print(f"üíæ Î™®Îç∏ Ï†ÄÏû• Í≤ΩÎ°ú: {MODEL_SAVE_DIR}")
```

# Î™®Îç∏ ÏÑ±Îä• ÌÖåÏä§Ìä∏

```python
# ==========================================
# 7. ÏÑ±Îä• ÌÖåÏä§Ìä∏ (Inference)
# ==========================================
def predict(text):
    model.eval() # ÌèâÍ∞Ä Î™®ÎìúÎ°ú Ï†ÑÌôò
    
    inputs = tokenizer(
        text, 
        return_tensors='pt', 
        truncation=True, 
        max_length=MAX_LEN, 
        padding='max_length'
    ).to(DEVICE)

    with torch.no_grad():
        outputs = model(**inputs)
        # ÌôïÎ•† Í≥ÑÏÇ∞ (Softmax)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
        pred_idx = torch.argmax(probs, dim=-1).item()
        confidence = probs[0][pred_idx].item()

    result = LABEL_MAP[pred_idx]
    print(f"üìù ÏûÖÎ†•: {text}")
    print(f"ü§ñ Í≤∞Í≥º: {result} (ÌôïÏã†ÎèÑ: {confidence*100:.1f}%)\n")

# --- ÌÖåÏä§Ìä∏ Î¨∏Ïû• ÏûÖÎ†• ---
print("--- [AI Í∞êÏ†ïÎ∂ÑÏÑù ÌÖåÏä§Ìä∏] ---")
predict("12Ïõî 25Ïùº ÌïÄÌÅ¨Ïä§Îäî Ïù¥ÎØ∏ ÎßàÍ∞êÏù¥Îùº Ïñ¥Î†µÏäµÎãàÎã§.")  # ÏòàÏÉÅ: Î∂àÍ∞Ä
predict("ÍπÄÏ≤†ÏàòÎãò Ìï≠Í≥µÍ∂å Î∞úÍ∂å ÏôÑÎ£åÌï¥ÎìúÎ†∏ÏäµÎãàÎã§.")        # ÏòàÏÉÅ: ÌôïÏ†ï
predict("ÌòÑÏßÄ Ìò∏ÌÖîÏóê ÏóëÏä§Ìä∏ÎùºÎ≤†Îìú Í∞ÄÎä•ÌïúÏßÄ ÌôïÏù∏ Ï§ëÏûÖÎãàÎã§.") # ÏòàÏÉÅ: Î≥¥Î•ò
predict("Ï£ºÎßêÏù¥Îùº ÏöîÍ∏àÏù¥ Ï¢Ä ÎπÑÏãºÎç∞ ÏßÑÌñâÌïòÏã§ÍπåÏöî?")       # Ïï†Îß§Ìïú Î¨∏Ïû• ÌÖåÏä§Ìä∏
predict("ÏÇ¨Ïû•Îãò Í∑∏ ÎÇ†ÏßúÎäî Ï†ÑÎ©∏Ïù¥Îùº Ïñ¥Î†µÍ≤†ÎäîÎç∞Ïöî")  # Î∂àÍ∞Ä ÏòàÏÉÅ
predict("ÏùºÎã® Ïû°Í≥† 30Î∂Ñ Îí§Ïóê Îã§Ïãú ÌÜ° ÎìúÎ¶¥Í≤åÏöî")    # Î≥¥Î•ò ÏòàÏÉÅ
predict("Ìã∞ÏºìÌåÖ ÎÅùÎÇ¨ÏäµÎãàÎã§ ÌååÏùº ÌôïÏù∏Ìï¥Ï£ºÏÑ∏Ïöî")      # ÌôïÏ†ï ÏòàÏÉÅ

print("--- [Ïã¨Ìôî Í≤ÄÏ¶ù ÌÖåÏä§Ìä∏: Í≥ºÏó∞ AIÎäî ÎòëÎòëÌï¥Ï°åÏùÑÍπå?] ---\n")

# üî¥ Ìå®ÌÑ¥ 1: "ÏùºÎã® Ïû°ÏïòÎã§"Îäî ÎßêÏóê ÏÜçÏßÄ ÎßàÎùº! (Ìï®Ï†ï: ÌôïÏ†ïÏ≤òÎüº Î≥¥Ïù¥ÏßÄÎßå Î≥¥Î•òÏûÑ)
# ÏïÑÍπå ÌãÄÎ¶∞ Î¨∏Ïû•: "ÏùºÎã® Ïû°Í≥† 30Î∂Ñ Îí§Ïóê Îã§Ïãú ÌÜ° ÎìúÎ¶¥Í≤åÏöî"
predict("Ï¢åÏÑùÏùÄ ÏùºÎã® ÎπºÎÜ®ÎäîÎç∞ ÏûÖÍ∏à ÌôïÏù∏Ïù¥ Ïïà ÎêòÎÑ§Ïöî")      # Ï†ïÎãµ: Î≥¥Î•ò (2)
predict("Í∞ÄÏòàÏïΩ Í±∏Ïñ¥Îëî Í±∞Îùº 5ÏãúÍπåÏßÄ Í≤∞Ï†ú ÏóÜÏúºÎ©¥ ÌíÄÎ¶ΩÎãàÎã§") # Ï†ïÎãµ: Î≥¥Î•ò (2)
predict("Î∞©ÏùÄ Ïû°ÏïÑÎíÄÎäîÎç∞ ÏïÑÏßÅ ÌôïÏ†ïÏÑúÍ∞Ä Ïïà ÎÇòÏôîÏñ¥Ïöî")       # Ï†ïÎãµ: Î≥¥Î•ò (2)

# üü† Ìå®ÌÑ¥ 2: "Î∂ÄÏ†ïÏ†Å Îã®Ïñ¥(ÎπÑÏã∏Îã§, Î©ÄÎã§)"Ïóê Ï´ÑÏßÄ ÎßàÎùº! (Ìï®Ï†ï: Í±∞Ï†àÏ≤òÎüº Î≥¥Ïù¥ÏßÄÎßå ÏßàÎ¨∏ÏûÑ)
# ÏïÑÍπå ÌãÄÎ¶∞ Î¨∏Ïû•: "Ï£ºÎßêÏù¥Îùº ÏöîÍ∏àÏù¥ Ï¢Ä ÎπÑÏãºÎç∞ ÏßÑÌñâÌïòÏã§ÍπåÏöî?"
predict("Ï∂îÏÑùÏù¥Îùº ÏöîÍ∏àÏù¥ ÌèâÏÜå 2Î∞∞Ïù∏Îç∞ Í¥úÏ∞ÆÏúºÏã†Í∞ÄÏöî?")      # Ï†ïÎãµ: Î≥¥Î•ò (2)
predict("ÏõêÌïòÏãúÎäî Î∑∞Îäî ÎßàÍ∞êÏù¥Îùº Í∞ÄÎì†Î∑∞Î°ú ÏßÑÌñâÌï†ÍπåÏöî?")     # Ï†ïÎãµ: Î≥¥Î•ò (2)

# üü° Ìå®ÌÑ¥ 3: Í∑πÎã®Ï†ÅÏù∏ Í±∞Ï†à ÌëúÌòÑ (Ïä¨Îû≠/Í¥ÄÏö©Íµ¨ ÌÖåÏä§Ìä∏)
# ÏïÑÍπå ÎßûÏ∂ò Î¨∏Ïû•: "ÏÇ¨Ïû•Îãò Í∑∏ ÎÇ†ÏßúÎäî Ï†ÑÎ©∏Ïù¥Îùº Ïñ¥Î†µÍ≤†ÎäîÎç∞Ïöî"
predict("Í∑∏ ÎÇ†ÏßúÎäî ÎåÄÍ∏∞ÎèÑ Ïïà Î∞õÏïÑÏ§Ñ Ï†ïÎèÑÎ°ú ÍΩâ ÎßâÌòîÏäµÎãàÎã§") # Ï†ïÎãµ: Î∂àÍ∞Ä (0)
predict("Ìã∞ÏóÖÏù¥ Ïî®Í∞Ä ÎßêÎùºÏÑú ÎèÑÏ†ÄÌûà Íµ¨Ìï† ÏàòÍ∞Ä ÏóÜÎÑ§Ïöî")      # Ï†ïÎãµ: Î∂àÍ∞Ä (0)
```

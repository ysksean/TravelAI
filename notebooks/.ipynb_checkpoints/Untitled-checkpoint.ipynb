```python
import os
import torch
import torch.nn as nn
import numpy as np
from transformers import (
    ElectraConfig, ElectraForTokenClassification, ElectraForSequenceClassification, # M1, M2μ©
    BartConfig, BartForConditionalGeneration, # M3μ©
    AutoTokenizer
)

# === μ„¤μ • ===
BASE_DIR = os.getcwd()
MODEL_DIR = os.path.join(BASE_DIR, "advanced_models_sota")

# ν΄λ” μƒμ„±
os.makedirs(MODEL_DIR, exist_ok=True)
print(f"π“‚ λ¨λΈ μ €μ¥ κ²½λ΅: {MODEL_DIR}")

# CPU/GPU μ„¤μ •
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"π”Ή Device: {device}")
```

```python
print("--- [M1] KoELECTRA (NER) μƒμ„± λ° μ €μ¥ ---")

# 1. μ„¤μ • (Config) μƒμ„± - μ‹¤μ  ν•™μµλ κ°€μ¤‘μΉ λ€μ‹  κµ¬μ΅°λ§ κ°€μ Έμ΄ (λΉ λ¥Έ ν…μ¤νΈμ©)
# μ‹¤μ  μ‚¬μ©μ‹λ” 'monologg/koelectra-base-v3-discriminator' λ“±μ„ μ‚¬μ©
config_m1 = ElectraConfig(
    vocab_size=30000,
    hidden_size=64, # ν…μ¤νΈμ©μ΄λΌ μ‘κ² μ„¤μ • (μ›λλ” 768)
    num_hidden_layers=2, # λ μ΄μ–΄ μλ„ μ¶•μ†
    num_labels=5, # νƒκ·Έ κ°μ (B-Hotel, I-Price λ“±)
    max_position_embeddings=128
)

# 2. λ¨λΈ μ΄κΈ°ν™” (Random Weights)
model_m1 = ElectraForTokenClassification(config_m1).to(device)

# 3. μ €μ¥ (Hugging Face ν‘μ¤€ λ°©μ‹)
m1_path = os.path.join(MODEL_DIR, "koelectra_ner")
model_m1.save_pretrained(m1_path)
print(f"β… λ¨λΈ μ €μ¥ μ™„λ£: {m1_path}")

# --- Load & Test ---
print("\n--- [M1] λ΅λ“ λ° μ¶”λ΅  ν…μ¤νΈ ---")
loaded_m1 = ElectraForTokenClassification.from_pretrained(m1_path).to(device)
loaded_m1.eval()

# λ”λ―Έ μ…λ ¥ (Batch: 1, Length: 50)
dummy_input_ids = torch.randint(0, 30000, (1, 50)).to(device)

with torch.no_grad():
    outputs = loaded_m1(dummy_input_ids)
    logits = outputs.logits # (1, 50, 5)

print(f"   γ„΄ μ¶λ ¥ ν•νƒ: {logits.shape}")
print("   γ„΄ μ •μƒ μ‘λ™ ν™•μΈ.")
```

```python
print("\n--- [M2] KoELECTRA Small (Sentiment) μƒμ„± λ° μ €μ¥ ---")

# 1. μ„¤μ • (3-Class: κΈμ •/λ¶€μ •/λ³΄λ¥)
config_m2 = ElectraConfig(
    vocab_size=30000,
    hidden_size=64,
    num_hidden_layers=2,
    num_labels=3, # 3μ¤‘ λ¶„λ¥
    max_position_embeddings=128
)

# 2. λ¨λΈ μ΄κΈ°ν™”
model_m2 = ElectraForSequenceClassification(config_m2).to(device)

# 3. μ €μ¥
m2_path = os.path.join(MODEL_DIR, "koelectra_sentiment")
model_m2.save_pretrained(m2_path)
print(f"β… λ¨λΈ μ €μ¥ μ™„λ£: {m2_path}")

# --- Load & Test ---
print("\n--- [M2] λ΅λ“ λ° μ¶”λ΅  ν…μ¤νΈ ---")
loaded_m2 = ElectraForSequenceClassification.from_pretrained(m2_path).to(device)
loaded_m2.eval()

dummy_input_ids = torch.randint(0, 30000, (1, 50)).to(device)

with torch.no_grad():
    outputs = loaded_m2(dummy_input_ids)
    probs = torch.softmax(outputs.logits, dim=-1)

print(f"   γ„΄ μμΈ΅ ν™•λ¥  λ¶„ν¬: {probs[0].cpu().numpy()}")
```

```python
print("\n--- [M3] KoBART (Summarizer) μƒμ„± λ° μ €μ¥ ---")

# 1. μ„¤μ • (Encoder-Decoder κµ¬μ΅°)
config_m3 = BartConfig(
    vocab_size=30000,
    d_model=64,       # hidden size
    encoder_layers=2,
    decoder_layers=2,
    max_position_embeddings=128
)

# 2. λ¨λΈ μ΄κΈ°ν™”
model_m3 = BartForConditionalGeneration(config_m3).to(device)

# 3. μ €μ¥
m3_path = os.path.join(MODEL_DIR, "kobart_summary")
model_m3.save_pretrained(m3_path)
print(f"β… λ¨λΈ μ €μ¥ μ™„λ£: {m3_path}")

# --- Load & Test ---
print("\n--- [M3] λ΅λ“ λ° μƒμ„± ν…μ¤νΈ ---")
loaded_m3 = BartForConditionalGeneration.from_pretrained(m3_path).to(device)
loaded_m3.eval()

dummy_input_ids = torch.randint(0, 30000, (1, 100)).to(device) # κΈ΄ λ¬Έμ¥ μ…λ ¥

with torch.no_grad():
    # generate() ν•¨μλ¥Ό ν†µν•΄ μ”μ•½λ¬Έ μƒμ„±
    summary_ids = loaded_m3.generate(dummy_input_ids, max_length=20)

print(f"   γ„΄ μƒμ„±λ μ”μ•½λ¬Έ ν† ν° ID: {summary_ids.shape}")
```

```python
print("\n--- [M4] N-BEATS Style (Forecaster) μƒμ„± λ° μ €μ¥ ---")

# 1. κ°„λ‹¨ν• N-BEATS μ¤νƒ€μΌ λ¨λΈ μ •μ (PyTorch)
class SimpleNBeats(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim=64):
        super(SimpleNBeats, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, output_dim) # Trend μμΈ΅
        self.relu = nn.ReLU()
        
    def forward(self, x):
        # x shape: (Batch, Time_Steps) - μ—¬κΈ°μ„  λ‹¨μν™”
        out = self.relu(self.fc1(x))
        out = self.relu(self.fc2(out))
        return self.fc3(out)

# 2. λ¨λΈ μ΄κΈ°ν™”
TIME_STEPS = 30 # κ³Όκ±° 30μΌ
FORECAST_HORIZON = 1 # λ‹¤μλ‚  μμΈ΅
model_m4 = SimpleNBeats(input_dim=TIME_STEPS, output_dim=FORECAST_HORIZON).to(device)

# 3. μ €μ¥ (PyTorch λ°©μ‹)
m4_path = os.path.join(MODEL_DIR, "nbeats_forecast.pth")
torch.save(model_m4.state_dict(), m4_path)
print(f"β… λ¨λΈ μ €μ¥ μ™„λ£: {m4_path}")

# --- Load & Test ---
print("\n--- [M4] λ΅λ“ λ° μμΈ΅ ν…μ¤νΈ ---")
loaded_m4 = SimpleNBeats(input_dim=TIME_STEPS, output_dim=FORECAST_HORIZON).to(device)
loaded_m4.load_state_dict(torch.load(m4_path))
loaded_m4.eval()

# λ”λ―Έ μ…λ ¥ (Batch: 1, 30μΌμΉ λ°μ΄ν„°)
dummy_input = torch.randn(1, TIME_STEPS).to(device)

with torch.no_grad():
    pred = loaded_m4(dummy_input)

print(f"   γ„΄ μ…λ ¥ λ°μ΄ν„°: {dummy_input.shape}")
print(f"   γ„΄ μμΈ΅κ°’: {pred.item()}")
```

```python
import os
import json
import google.generativeai as genai
from pdf2image import convert_from_path
from PIL import Image
from typing import Dict, Any, List

# ==========================================
# [μ„¤μ •]
# ==========================================
MY_API_KEY = "AIzaSyBRQJ1uVE5UVvB6jO6HvRhB5LTfWncsqN8"  # λ³ΈμΈμ API ν‚¤
MODEL_NAME = "gemini-2.5-flash"
genai.configure(api_key=MY_API_KEY, transport="rest")

# β… μ¤‘μ”: μ•„κΉ λ‹¤μ΄λ°›μ€ Popplerμ bin ν΄λ” κ²½λ΅λ¥Ό μ—¬κΈ°μ— λ¶™μ—¬λ„£μΌμ„Έμ” ( \ λ€μ‹  / μ‚¬μ© κ¶μ¥ )
# μμ‹: r"C:\Start\poppler-24.02.0\Library\bin"
POPPLER_BIN_PATH = r"C:\poppler\Library\bin"

class TravelVisionAI:
    def __init__(self):
        self.model = genai.GenerativeModel(MODEL_NAME)
        # Target Schema
        self.target_schema = {
            "basic_info": {"product_type": None, "is_flight_included": None, "is_vat_included": None},
            "location_info": {"country": None, "city": None, "departure_port": None},
            "product_info": {
                "product_name": None,
                "event_period": {"available_days": []}
            },
            "hotels": [{"name_kr": None, "description": None, "meta_info": {"check_in_out": None, "website": None}}],
            "golf_courses": [{"name_kr": None, "operation_info": None, "meta_info": {"detail_info": None, "website": None}}],
            "tourist_spots": [{"name": None}],
            "details": {
                "inclusions": [], "exclusions": [], "others": None,
                "is_insurance_included": None, "is_guide_included": None,
                "special_notes": []
            },
            "ai_content": {"body_text": None},
            "flight_info": {"airline": None, "flight_number": None, "departure_time": None, "arrival_time": None},
            "price_info": []
        }

    def _convert_to_pdf_if_needed(self, file_path):
        """
        μ…λ ¥ νμΌμ΄ DOCXλΌλ©΄ PDFλ΅ λ³€ν™ν•μ—¬ κ²½λ΅ λ°ν™.
        μ΄λ―Έ PDFλΌλ©΄ κ·Έλ€λ΅ λ°ν™.
        """
        ext = os.path.splitext(file_path)[1].lower()

        # 1. μ΄λ―Έ PDFμΈ κ²½μ°
        if ext == '.pdf':
            return file_path

        # 2. DOCXμΈ κ²½μ° -> PDF λ³€ν™
        elif ext in ['.docx', '.doc']:
            print(f"   β„ΉοΈ DOCX κ°μ§€λ¨. PDFλ΅ λ³€ν™ μ¤‘... ({os.path.basename(file_path)})")
            if docx_to_pdf_tool is None:
                print("   β docx2pdf λΌμ΄λΈλ¬λ¦¬κ°€ μ„¤μΉλμ§€ μ•μ•μµλ‹λ‹¤. (!pip install docx2pdf)")
                return None

            # κ°™μ€ ν΄λ”μ— κ°™μ€ μ΄λ¦„μ pdf μƒμ„±
            output_pdf = os.path.splitext(file_path)[0] + ".pdf"
            try:
                docx_to_pdf_tool(file_path, output_pdf)
                print(f"   β… λ³€ν™ μ™„λ£: {output_pdf}")
                return output_pdf
            except Exception as e:
                print(f"   β DOCX λ³€ν™ μ‹¤ν¨ (MS Wordκ°€ μ„¤μΉλμ–΄ μμ–΄μ•Ό ν•¨): {e}")
                return None

        # 3. μ§€μ›ν•μ§€ μ•λ” νμΌ
        else:
            print(f"   β μ§€μ›ν•μ§€ μ•λ” νμΌ ν•μ‹μ…λ‹λ‹¤: {ext}")
            return None

    def _file_to_images(self, file_path):
        """νμΌ κ²½λ΅λ¥Ό λ°›μ•„ (λ³€ν™ ν›„) μ΄λ―Έμ§€ λ¦¬μ¤νΈ λ¦¬ν„΄"""

        # 1. PDFκ°€ μ•„λ‹λΌλ©΄ PDFλ΅ λ³€ν™
        pdf_path = self._convert_to_pdf_if_needed(file_path)

        if not pdf_path or not os.path.exists(pdf_path):
            return []

        print(f"   -> μ΄λ―Έμ§€ λ³€ν™ μ¤‘: {os.path.basename(pdf_path)}")
        try:
            return convert_from_path(pdf_path, dpi=400, poppler_path=POPPLER_BIN_PATH)
        except Exception as e:
            print(f"β Poppler μ—λ¬: {e}")
            return []

    def process(self, product_file: str, price_file: str = None):
        print(">>> [1/3] λ¬Έμ„ μ²λ¦¬ μ‹μ‘...")

        # [μμ •] _pdf_to_images λ€μ‹  _file_to_images μ‚¬μ© (DOCX μλ™ μ²λ¦¬)
        prod_images = self._file_to_images(product_file)
        price_images = self._file_to_images(price_file) if price_file else []

        if not prod_images:
            return {"error": "μƒν’ νμΌ λ³€ν™ μ‹¤ν¨"}

        print(">>> [2/3] Gemini Vision λ¶„μ„ μ¤‘...")

        # 1. μƒν’ μ •λ³΄ μ¶”μ¶
        product_json = self._extract_product_info(prod_images)

        # 2. κ°€κ²© μ •λ³΄ μ¶”μ¶
        if price_images:
            print("   -> [Mode] λ³„λ„ κ°€κ²© νμΌ λ¶„μ„")
            target_images = price_images
        else:
            print("   -> [Mode] ν†µν•© νμΌ λ¶„μ„")
            target_images = prod_images

        price_json = self._extract_price_info(target_images)

        print(">>> [3/3] λ°μ΄ν„° λ³‘ν•© μ¤‘...")
        final_data = self._merge(product_json, price_json)
        return final_data

    def _extract_product_info(self, images):
        # [μμ •] product_type λ¶„λ¥ λ΅μ§ μ¶”κ°€ (Prompt + Python Logic)
        prompt = f"""
        Role: Travel Product Analyst.
        Task: Analyze the images and extract product details into JSON.

        [CRITICAL LANGUAGE RULES]
        1. **OUTPUT VALUES MUST BE IN KOREAN.** (e.g., "Japan"->"μΌλ³Έ")
        2. **DO NOT TRANSLATE KEYS.**

        [Extraction Focus]
        1. **Location**: Identify the `country` and `city` accurately.
           - If `country` is South Korea, map as "ν•κµ­".
        2. **Product Type Logic**:
           - If `country` is "ν•κµ­" (or Jeju), set `product_type` to "κµ­λ‚΄μƒν’".
           - Otherwise, set `product_type` to "ν•΄μ™Έμƒν’".
        3. **Flight Info**: Look specifically for flight codes (LJ, KE, TW, etc.) and times.
        4. Flight Status: Set is_flight_included to "ν¬ν•¨" if flight codes are present, otherwise "λ¶ν¬ν•¨".

        [Target JSON Schema]
        {json.dumps(self.target_schema, ensure_ascii=False)}
        """
        try:
            # μƒν’ μ •λ³΄λ” μ†λ„λ¥Ό μ„ν•΄ μ•λ¶€λ¶„ 5μ¥λ§ ν™•μΈ (μ„ νƒμ‚¬ν•­)
            content = [prompt] + images[:5]
            res = self.model.generate_content(
                content,
                generation_config={"response_mime_type": "application/json"},
                request_options={"timeout": 120}
            )
            data = json.loads(res.text)

            if isinstance(data, list):
                data = data[0] if len(data) > 0 else {}

            # ==========================================================
            # [Python ν›„μ²λ¦¬] product_type κ°•μ  λ³΄μ • λ΅μ§
            # ==========================================================
            country = data.get("location_info", {}).get("country", "")
            if country:
                if any(x in country for x in ["ν•κµ­", "λ€ν•λ―Όκµ­", "μ μ£Ό", "Korea"]):
                    data["basic_info"]["product_type"] = "κµ­λ‚΄μƒν’"
                    data["location_info"]["country"] = "ν•κµ­"
                else:
                    data["basic_info"]["product_type"] = "ν•΄μ™Έμƒν’"

            return data

        except Exception as e:
            print(f"Product Extract Error: {e}")
            return {}

    def _extract_price_info(self, images):
        all_prices = []
        seen_entries = set()

        print(f"   -> μ΄ {len(images)}μ¥μ κ°€κ²©ν‘ μ΄λ―Έμ§€λ¥Ό λ¶„μ„ν•©λ‹λ‹¤ (ν•μ΄λΈλ¦¬λ“ μ¤μΊ”).")

        for page_idx, full_img in enumerate(images):
            width, height = full_img.size

            # [μ¶”κ°€] ν•μ΄λΈλ¦¬λ“ μ²­ν¬ μƒμ„±μ „λµ
            # μ „λµ 1: μ „μ²΄ μ΄λ―Έμ§€ ν†µμ§Έλ΅ μ¤μΊ” (μΌλ°μ μΈ ν‘ λ€μ‘, idx=-1λ΅ κµ¬λ¶„)
            chunks = [(full_img, -1, "Full Page")]

            # μ „λµ 2: μμ§ 3λ“±λ¶„ μ¤μΊ” (λ³µμ΅ν• μ—‘μ…€ λ¤ν”„ λ€μ‘)
            # μ΄λ―Έμ§€κ°€ μ¶©λ¶„ν λ„“μ„ λ•λ§ μν–‰ (μ: λ„λΉ„ 1000px μ΄μƒ)
            if width > 1000:
                split_1 = int(width / 3)
                split_2 = int(width * 2 / 3)
                chunks.append((full_img.crop((0, 0, split_1, height)), 0, "Left Column"))
                chunks.append((full_img.crop((split_1, 0, split_2, height)), 1, "Middle Column"))
                chunks.append((full_img.crop((split_2, 0, width, height)), 2, "Right Column"))

            print(f"      Running Page {page_idx + 1} ({len(chunks)} Scans)...")

            for img_chunk, col_idx, position in chunks:

                # [ν•µμ‹¬ λ΅μ§ 1] μ»¬λΌ μ„μΉμ— λ”°λ¥Έ κΈ°λ³Έκ°’ μ„¤μ •
                if col_idx == 0:
                    def_night, def_day, def_group = 2, 3, 4
                elif col_idx == 1:
                    def_night, def_day, def_group = 3, 4, 4
                elif col_idx == 2:
                    def_night, def_day, def_group = 4, 5, 4
                else:
                    # [μ¶”κ°€] col_idx == -1 (Full Page)μΌ λ•μ κΈ°λ³Έκ°’
                    # μΌλ°μ μΈ μƒν’μΌ κ²½μ° κ°€μ¥ ν”ν• ν¨ν„΄(3λ°•4μΌ, 4μΈ)μ„ κΈ°λ³ΈμΌλ΅ μ΅κ±°λ‚,
                    # μ΄ κ°’μ€ μ‹¤μ  νμ‹±λ  λ• λ¬΄μ‹λ  μλ„ μμµλ‹λ‹¤.
                    def_night, def_day, def_group = 3, 4, 4

                prompt = f"""
                Role: Travel Price Analyst.
                Task: Extract pricing rows from this image fragment ({position}).

                [CRITICAL INSTRUCTION: CSV FORMAT]
                1. Output **CSV ONLY**. No JSON, No Markdown.
                2. Format: `departure_date, price_adult`
                3. **Ignore missing columns**: I will fill night/day/group programmatically.
                4. **Date Parsing**: Convert "12μ›” 01μΌ" -> "2025-12-01" (Assume 2025 for Dec, 2026 for Jan+).
                5. **Price Parsing**: Extract numbers only (e.g., 1,010,000 -> 1010000). Skip empty rows.
                """

                try:
                    res = self.model.generate_content(
                        [prompt, img_chunk],
                        generation_config={
                            "max_output_tokens": 8192,
                            "temperature": 0.0
                        },
                        request_options={"timeout": 120}
                    )

                    csv_text = res.text.strip()

                    # [ν•µμ‹¬ λ΅μ§ 2] νμ„μ— κΈ°λ³Έκ°’(defaults)μ„ μ „λ‹¬ν•μ—¬ κ°•μ  μ£Όμ…
                    new_rows = self._parse_csv_to_json(
                        csv_text,
                        seen_entries,
                        defaults=(def_night, def_day, def_group)
                    )

                    if new_rows:
                        all_prices.extend(new_rows)
                        # print(f"         β… {position}: Extracted {len(new_rows)} rows.")
                    # else:
                        # print(f"         β οΈ {position}: No data found.")

                except Exception as e:
                    # print(f"         β {position} Error: {e}")
                    continue

        # λ‚ μ§μ μ •λ ¬
        all_prices.sort(key=lambda x: x['departure_date'])

        return {
            "global_updates": {"is_vat_included": None, "cancellation_refund": None},
            "price_info": all_prices
        }

    def _parse_csv_to_json(self, csv_text, seen_entries, defaults):
        """CSVλ¥Ό νμ‹±ν•κ³ , λ‚ μ§/κ°€κ²© μ™Έμ μ •λ³΄λ” μ „λ‹¬λ°›μ€ defaults κ°’μΌλ΅ μ±„μ›€"""
        rows = []
        lines = csv_text.split('\n')

        # μ „λ‹¬λ°›μ€ κ°•μ  μ£Όμ… κ°’λ“¤
        def_night, def_day, def_group = defaults

        for line in lines:
            line = line.strip()
            if not line: continue
            if "date" in line.lower() or "price" in line.lower(): continue

            parts = line.split(',')
            # ν”„λ΅¬ν”„νΈμ—μ„ λ‚ μ§, κ°€κ²©λ§ λ‹¬λΌκ³  ν–μΌλ―€λ΅ 2κ° μ΄μƒμ΄λ©΄ λ¨
            if len(parts) < 2: continue

            try:
                # 1. λ‚ μ§ μ •κ·ν™”
                raw_date = parts[0].strip()
                clean_date = raw_date.replace('.', '-').replace('/', '-')

                # λ‚ μ§μ— ν•κΈ€μ΄λ‚ κ³µλ°±μ΄ μ„μΈ κ²½μ° μ²λ¦¬ (μ: "12μ›” 01μΌ")
                import re
                nums = re.findall(r'\d+', clean_date)
                if len(nums) >= 2:
                    month = int(nums[-2])
                    day = int(nums[-1])

                    # μ—°λ„ μ¶”μ • (10,11,12μ›”μ€ 2025λ…„, 1,2,3μ›”μ€ 2026λ…„μΌλ΅ κ°€μ •)
                    year = 2025 if month >= 10 else 2026
                    clean_date = f"{year}-{month:02d}-{day:02d}"
                else:
                    continue # λ‚ μ§ μΈμ‹ μ‹¤ν¨ μ‹ μ¤ν‚µ

                # 2. κ°€κ²© μ¶”μ¶ (λ§μ§€λ§‰ μ»¬λΌμ΄ κ°€κ²©μΌ ν™•λ¥  λ†’μ)
                price_str = parts[-1].strip()
                price = int(re.sub(r'\D', '', price_str))

                # 3. μ¤‘λ³µ μ²΄ν¬ ν‚¤
                unique_key = f"{clean_date}_{def_night}_{def_day}_{def_group}_{price}"

                if unique_key not in seen_entries:
                    seen_entries.add(unique_key)

                    rows.append({
                        "departure_date": clean_date,
                        "night_count": def_night,  # κ°•μ  μ£Όμ…
                        "day_count": def_day,      # κ°•μ  μ£Όμ…
                        "group_size": def_group,   # κ°•μ  μ£Όμ…
                        "price_adult": price,
                        "status": "available"
                    })
            except Exception:
                continue

        return rows

    def _merge(self, prod, price):
        # λ°©μ–΄ μ½”λ“: prodκ°€ μ—¬μ „ν dictκ°€ μ•„λ‹λ©΄ λΉ dictλ΅ μ΄κΈ°ν™”
        if not isinstance(prod, dict): prod = {}
        if not isinstance(price, dict): price = {}

        # 1. κ°€κ²© λ¦¬μ¤νΈ λ³‘ν•©
        prod["price_info"] = price.get("price_info", [])

        # 2. κΈ€λ΅λ² μ •μ±… μ—…λ°μ΄νΈ
        updates = price.get("global_updates", {})

        # basic_info ν‚¤κ°€ μ—†μΌλ©΄ μƒμ„± (λ°©μ–΄ μ½”λ“)
        if "basic_info" not in prod: prod["basic_info"] = {}
        if "policies" not in prod: prod["policies"] = {}

        if updates.get("is_vat_included") is not None:
            prod["basic_info"]["is_vat_included"] = updates["is_vat_included"]

        if updates.get("cancellation_refund"):
            prod["policies"]["cancellation_refund"] = updates["cancellation_refund"]

        return prod

# --- μ‹¤ν–‰ ---
if __name__ == "__main__":
    # νμΌ κ²½λ΅ μ„¤μ • (ν…μ¤νΈμ©)
    PRODUCT_FILE = "../ERP ν•„μ”ν• λ°μ΄ν„°/1. λλ“μ‚¬ν•ν… λ°›μ€ μƒν’_μ™„λ£/μƒν’/test.pdf"
    PRICE_FILE = "../ERP ν•„μ”ν• λ°μ΄ν„°/1. λλ“μ‚¬ν•ν… λ°›μ€ μƒν’_μ™„λ£/κ°€κ²©ν‘/text.pdf"

    if os.path.exists(PRODUCT_FILE):
        ai = TravelVisionAI()

        # Case 1: μƒν’+κ°€κ²© νμΌ λ‘ λ‹¤ μμ„ λ•
        if os.path.exists(PRICE_FILE):
            print(">>> [TEST] μƒν’ + κ°€κ²© νμΌ λ¨λ“")
            result = ai.process(PRODUCT_FILE, PRICE_FILE)

        # Case 2: μƒν’ νμΌλ§ μμ„ λ• (μƒν’ νμΌ λ‚΄μ— κ°€κ²©ν‘κ°€ μλ‹¤κ³  κ°€μ •)
        else:
            print(">>> [TEST] ν†µν•© νμΌ λ¨λ“ (κ°€κ²© νμΌ μ—†μ)")
            result = ai.process(PRODUCT_FILE, None)

        print("\n" + "="*50)
        print("π― FINAL JSON RESULT")
        print("="*50)
        print(json.dumps(result, indent=2, ensure_ascii=False))
    else:
        print("β μƒν’ νμΌ κ²½λ΅λ¥Ό ν™•μΈν•΄μ£Όμ„Έμ”.")
```

```python
import os
import json
import google.generativeai as genai
from pdf2image import convert_from_path
from PIL import Image
from typing import Dict, Any, List
from typing_extensions import TypedDict
import typing
import pdfplumber
import pandas as pd

# ==========================================
# [μ„¤μ •]
# ==========================================
MY_API_KEY = "AIzaSyBRQJ1uVE5UVvB6jO6HvRhB5LTfWncsqN8"  # λ³ΈμΈμ API ν‚¤
MODEL_NAME = "gemini-2.5-flash-lite"
genai.configure(api_key=MY_API_KEY, transport="rest")

# β… μ¤‘μ”: μ•„κΉ λ‹¤μ΄λ°›μ€ Popplerμ bin ν΄λ” κ²½λ΅λ¥Ό μ—¬κΈ°μ— λ¶™μ—¬λ„£μΌμ„Έμ” ( \ λ€μ‹  / μ‚¬μ© κ¶μ¥ )
# μμ‹: r"C:\Start\poppler-24.02.0\Library\bin"
POPPLER_BIN_PATH = r"C:\poppler\Library\bin"

try:
    from docx2pdf import convert as docx_to_pdf_tool
except ImportError:
    docx_to_pdf_tool = None

# ==========================================
# [μ¤ν‚¤λ§ μ •μ] μ—¬ν–‰ μƒν’ λ°μ΄ν„° κµ¬μ΅°
# ==========================================

class BasicInfo(TypedDict):
    product_type: str | None       # κµ­λ‚΄μƒν’ / ν•΄μ™Έμƒν’
    is_flight_included: bool | None
    is_vat_included: bool | None

class LocationInfo(TypedDict):
    country: str | None
    city: str | None
    departure_port: str | None

class EventPeriod(TypedDict):
    available_days: list[str]

class ProductInfo(TypedDict):
    product_name: str | None
    event_period: EventPeriod

class MetaInfoHotel(TypedDict):
    check_in_out: str | None
    website: str | None

class Hotel(TypedDict):
    name_kr: str | None
    description: str | None        # λ“±κΈ‰(3μ„±/4μ„±) μ •λ³΄λ‚ μ„μΉ μ„¤λ… λ“±
    meta_info: MetaInfoHotel

class MetaInfoGolf(TypedDict):
    detail_info: str | None
    website: str | None

class GolfCourse(TypedDict):
    name_kr: str | None
    operation_info: str | None
    meta_info: MetaInfoGolf

class TouristSpot(TypedDict):
    name: str | None

class Details(TypedDict):
    inclusions: list[str]          # ν¬ν•¨μ‚¬ν•­
    exclusions: list[str]          # λ¶ν¬ν•¨μ‚¬ν•­
    others: str | None             # κΈ°νƒ€
    is_insurance_included: bool | None
    is_guide_included: bool | None
    special_notes: list[str]       # μ‡Όν•‘, μ„ νƒκ΄€κ΄‘, μ μμ‚¬ν•­

class AiContent(TypedDict):
    body_text: str | None          # AIκ°€ μ‘μ„±ν• λ§μΌ€ν…μ© μ”μ•½ λ³Έλ¬Έ

class FlightInfo(TypedDict):
    airline: str | None
    flight_number: str | None      # LJ357 λ“±
    departure_time: str | None
    arrival_time: str | None

class PriceInfo(TypedDict):
    departure_date: str
    night_count: int
    day_count: int
    group_size: int
    price_adult: int
    status: str

# β… λ©”μΈ μ¤ν‚¤λ§ (μµμΆ… μ¶λ ¥)
class TravelProductSchema(TypedDict):
    basic_info: BasicInfo
    location_info: LocationInfo
    product_info: ProductInfo
    hotels: list[Hotel]             # β… ν•µμ‹¬: λ¦¬μ¤νΈλ΅ μ •μ (1κ°λ“  3κ°λ“  λ‹¤ μμ©)
    golf_courses: list[GolfCourse]  # β… ν•µμ‹¬: λ¦¬μ¤νΈλ΅ μ •μ
    tourist_spots: list[TouristSpot]
    details: Details
    ai_content: AiContent
    flight_info: FlightInfo
    price_info: list[PriceInfo]

# κ°€κ²©ν‘ μ „μ© μ¤ν‚¤λ§
class PriceListSchema(TypedDict):
    prices: list[PriceInfo]

# ==========================================
# [ν΄λμ¤] λ§λ¥ μ—¬ν–‰ λ°μ΄ν„° μ²λ¦¬κΈ°
# ==========================================
class UniversalTravelAI:
    def __init__(self):
        self.model = genai.GenerativeModel(MODEL_NAME)

    # 429 μ—λ¬ λ°©μ§€μ© μ¬μ‹λ„ λ΅μ§
    def _generate_with_retry(self, content, config, retries=3):
        for i in range(retries):
            try:
                return self.model.generate_content(
                    content, generation_config=config, request_options={"timeout": 120}
                )
            except Exception as e:
                # ν• λ‹Ήλ‰ μ΄κ³Ό μ‹ λ€κΈ°
                if "429" in str(e) or "quota" in str(e).lower():
                    wait_time = (i + 1) * 15
                    print(f"   β οΈ Quota Exceeded. Retrying in {wait_time}s...")
                    time.sleep(wait_time)
                else: raise e
        return None

    # ---------------------------------------------------------
    # 1. ν…μ¤νΈ μ¶”μ¶κΈ° (λ¨λ“  νμΌ -> ν…μ¤νΈ)
    # ---------------------------------------------------------
    def _extract_text_content(self, file_path):
        ext = os.path.splitext(file_path)[1].lower()
        print(f"   π“„ [Text Extractor] Reading {ext} file...")
        try:
            if ext == '.pdf':
                text = ""
                with pdfplumber.open(file_path) as pdf:
                    for page in pdf.pages:
                        page_text = page.extract_text()
                        if page_text: text += page_text + "\n"
                return text
            elif ext in ['.docx', '.doc']:
                doc = Document(file_path)
                return "\n".join([p.text for p in doc.paragraphs])
            elif ext in ['.xlsx', '.xls']:
                xls = pd.read_excel(file_path, sheet_name=None)
                text = ""
                for sheet, df in xls.items():
                    text += f"--- Sheet: {sheet} ---\n"
                    # tabulate μ—†μΌλ©΄ to_stringμΌλ΅ λ€μ²΄ (μ—λ¬ λ°©μ§€)
                    try: text += df.to_markdown(index=False) + "\n"
                    except: text += df.to_string(index=False) + "\n"
                return text
            elif ext == '.txt':
                with open(file_path, 'r', encoding='utf-8') as f: return f.read()
            return ""
        except Exception as e:
            print(f"   β Text Extraction Failed: {e}")
            return ""

    # ---------------------------------------------------------
    # 2. μ΄λ―Έμ§€ λ³€ν™κΈ° (Visionμ©)
    # ---------------------------------------------------------
    def _convert_to_images(self, file_path):
        ext = os.path.splitext(file_path)[1].lower()
        print(f"   π–ΌοΈ [Image Converter] Processing {ext} file...")

        pdf_path = file_path
        # Docx -> PDF λ³€ν™
        if ext in ['.docx', '.doc']:
            try:
                from docx2pdf import convert
                pdf_path = os.path.splitext(file_path)[0] + ".pdf"
                convert(file_path, pdf_path)
            except: return []

        # PDF -> Images
        if pdf_path.lower().endswith('.pdf'):
            try:
                # Docker/Linux ν™κ²½μ—μ„λ” poppler_path μ κ±° ν•„μ”
                return convert_from_path(pdf_path, dpi=300, poppler_path=POPPLER_BIN_PATH)
            except: return []
        return []

    # ---------------------------------------------------------
    # 3. λ©”μΈ ν”„λ΅μ„Έμ¤ (μλ™ λ¶„κΈ° μ²λ¦¬)
    # ---------------------------------------------------------
    def analyze(self, product_file, price_file=None):
        print(f"\nπ€ Analysis Started")

        # [Step 1] μƒν’ μ •λ³΄ μ¶”μ¶ (ν…μ¤νΈ λ¨λ“)
        # ν…μ¤νΈ λ¨λ“λ” ν† ν°μ΄ μ €λ ΄ν•κ³ , URL/μ „ν™”λ²νΈ/νΉμλ¬Έμ μΈμ‹λ¥ μ΄ 100%μ…λ‹λ‹¤.
        product_text = self._extract_text_content(product_file)

        if not product_text:
            return {"error": "μƒν’ ν…μ¤νΈ μ¶”μ¶ μ‹¤ν¨"}

        # Gemini νΈμ¶
        product_data = self._call_gemini_product(product_text)

        # [Step 2] κ°€κ²© μ •λ³΄ μ¶”μ¶
        # κ°€κ²© νμΌμ΄ μ—†μΌλ©΄ μƒν’ νμΌμ„ κ°€κ²© λ¶„μ„ λ€μƒμΌλ΅ μ‚Όμ
        price_source = price_file if price_file else product_file
        price_ext = os.path.splitext(price_source)[1].lower()
        price_list = []

        # μ „λµ A: μ—‘μ…€/ν…μ¤νΈ νμΌ -> ν…μ¤νΈ λ¨λ“λ΅ λ¶„μ„ (ν‘ κµ¬μ΅° μ μ§€λ¨)
        if price_ext in ['.xlsx', '.xls', '.csv', '.txt']:
            print("   -> [Strategy A] Excel/Text detected.")
            raw_text = self._extract_text_content(price_source)
            price_list = self._call_gemini_price_text(raw_text)

        # μ „λµ B: PDF/μ΄λ―Έμ§€ νμΌ -> Vision λ¨λ“λ΅ λ¶„μ„ (λ³µμ΅ν• λ μ΄μ•„μ›ƒ μΈμ‹)
        else:
            print("   -> [Strategy B] PDF/Image detected.")
            images = self._convert_to_images(price_source)
            if images:
                price_list = self._call_gemini_price_vision(images)
            else:
                # μ΄λ―Έμ§€ λ³€ν™ μ‹¤ν¨ μ‹ ν…μ¤νΈ λ¨λ“λ΅ λΉ„μƒ μ „ν™
                print("   β οΈ Vision failed. Fallback to Text Analysis.")
                raw_text = self._extract_text_content(price_source)
                price_list = self._call_gemini_price_text(raw_text)

        # [Step 3] λ³‘ν•©
        if not product_data: product_data = {}
        product_data['price_info'] = price_list

        return product_data

    # ---------------------------------------------------------
    # 4. Gemini νΈμ¶ λ©”μ„λ“ (ν”„λ΅¬ν”„νΈ κ°•ν™”)
    # ---------------------------------------------------------
    def _call_gemini_product(self, text):
        """μƒν’ μƒμ„Έ μ •λ³΄ μ¶”μ¶ (λ‹¨μΌ/λ‹¤μ¤‘ μµμ… λ¨λ‘ λ€μ‘)"""

        prompt = """
        You are a generic travel product data parser.
        Analyze the text and extract details into the JSON schema perfectly.

        [MAPPING RULES - CRITICAL]

        1. **Multiple Options Handling (Hotels/Golf)**:
           - If the text lists multiple options (e.g., "3-star: A Hotel, 4-star: B Hotel" or "A CC, B CC, C CC"),
             **extract ALL of them** as separate items in the `hotels` or `golf_courses` list.
           - Do not merge them into one string. Create a list of objects.

        2. **flight_info (ν•­κ³µ)**:
           - Look for flight codes (e.g., LJ357, 7C, KE, OZ) and times.
           - If found, fill `flight_info` and set `is_flight_included` = True.

        3. **Location**:
           - If text contains "μ¤ν‚¤λ‚μ™€" or "λ―Έμ•Όμ½”μ§€λ§", set Country="μΌλ³Έ".

        4. **AI Content Creation**:
           - Read the whole text and write a summarizing marketing text in `ai_content.body_text`.
           - It should highlight key selling points (e.g., hotel grade, golf course view).

        5. **Language**:
           - All text output MUST be in **Korean**.
        """

        try:
            resp = self._generate_with_retry(
                [prompt, text[:50000]],
                genai.GenerationConfig(
                    response_mime_type="application/json",
                    response_schema=TravelProductSchema
                )
            )
            if resp:
                data = json.loads(resp.text)

                # ν›„μ²λ¦¬: κµ­λ‚΄/ν•΄μ™Έ μλ™ λ³΄μ •
                country = data.get("location_info", {}).get("country", "")
                if country and any(x in country for x in ["ν•κµ­", "λ€ν•λ―Όκµ­", "μ μ£Ό", "Korea"]):
                    data["basic_info"]["product_type"] = "κµ­λ‚΄μƒν’"
                else:
                    data["basic_info"]["product_type"] = "ν•΄μ™Έμƒν’"
                return data
            return {}
        except Exception as e:
            print(f"   β Product Info Error: {e}")
            return {}

    def _call_gemini_price_text(self, text):
        """κ°€κ²© μ •λ³΄ (Text λ¨λ“)"""
        prompt = "Extract pricing table. Rows: Date, Price, Headcount. Date format: YYYY-MM-DD."
        try:
            resp = self._generate_with_retry(
                [prompt, text[:30000]],
                genai.GenerationConfig(
                    response_mime_type="application/json",
                    response_schema=PriceListSchema
                )
            )
            return json.loads(resp.text).get('prices', []) if resp else []
        except Exception: return []

    def _call_gemini_price_vision(self, images):
        """κ°€κ²© μ •λ³΄ (Vision λ¨λ“)"""
        all_prices = []
        # ν† ν° μ μ•½μ„ μ„ν•΄ μ• 5μ¥λ§ λ¶„μ„ (ν•„μ”μ‹ μ΅°μ • κ°€λ¥)
        for img in images[:5]:
            try:
                time.sleep(2) # Rate Limit λ°©μ§€
                resp = self._generate_with_retry(
                    ["Extract price table. Output JSON only.", img],
                    genai.GenerationConfig(
                        response_mime_type="application/json",
                        response_schema=PriceListSchema
                    )
                )
                if resp:
                    items = json.loads(resp.text).get('prices', [])
                    all_prices.extend(items)
            except: continue
        return all_prices
```

```python
import json
import os
import sys
import torch
import numpy as np
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, ElectraForTokenClassification
from torch.optim import AdamW

# ====================================================
# 1. ν™κ²½ λ° κ²½λ΅ μ„¤μ •
# ====================================================
# ν„μ¬ λ””λ ‰ν† λ¦¬(travel_erp)μ μƒμ„ λ””λ ‰ν† λ¦¬(flask_web)λ¥Ό κ²½λ΅μ— μ¶”κ°€
flask_web_root = os.path.dirname(os.path.dirname(os.path.abspath(__file__))) if '__file__' in globals() else os.path.dirname(os.getcwd())
sys.path.append(flask_web_root) 
from services.parsing_service import parsing_manager

# μ„¤μ •
EPOCHS = 30              # μ¶©λ¶„ν ν•™μµν•λ„λ΅ λλ¦Ό
LEARNING_RATE = 5e-5
BATCH_SIZE = 4
MAX_LEN = 128
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

BASE_DIR = os.getcwd()
TRAIN_FILE = os.path.join(BASE_DIR, 'train_data.json')
TEST_FILE = '../ERP ν•„μ”ν• λ°μ΄ν„°/1. λλ“μ‚¬ν•ν… λ°›μ€ μƒν’_μ™„λ£/μƒν’/12μ›” ν•μ •νΉκ°€_ TW μ‚°μ•Ό μ§ν•­(μΈμ²) λ™λ°©λ©κ°€ CC,κ°•λ½μ›CC 27ν™€ λΌμ΄λ“ 5μΌ,6μΌ _ ν”λμ—μ΄ν¬μ–΄_251126.xlsx'

# νƒκ·Έ λ¦¬μ¤νΈ
LABEL_LIST = [
    "O",
    "B-HOTEL_NAME", "I-HOTEL_NAME", "B-HOTEL_GRADE", "I-HOTEL_GRADE", "B-HOTEL_LOC", "I-HOTEL_LOC",
    "B-GOLF_NAME", "I-GOLF_NAME", "B-GOLF_OP", "I-GOLF_OP",
    "B-FLIGHT_NAME", "I-FLIGHT_NAME", "B-FLIGHT_NUM", "I-FLIGHT_NUM", "B-DEPART_TIME", "I-DEPART_TIME",
    "B-PRICE", "I-PRICE", "B-INCLUSION", "I-INCLUSION", "B-EXCLUSION", "I-EXCLUSION",
    "B-REFUND", "I-REFUND", "B-DATE", "I-DATE", "B-CITY", "I-CITY", "B-NOTE", "I-NOTE",
    "B-PRODUCT_NAME", "I-PRODUCT_NAME", "B-DEPART_PORT", "I-DEPART_PORT", "B-PERIOD", "I-PERIOD"
]
LABEL2ID = {label: i for i, label in enumerate(LABEL_LIST)}
ID2LABEL = {i: label for i, label in enumerate(LABEL_LIST)}

# ====================================================
# 2. λ°μ΄ν„°μ…‹ ν΄λμ¤ (κ°•μ  ν¨λ”© λ΅μ§ μ¶”κ°€)
# ====================================================
class NERDataset(Dataset):
    def __init__(self, data, tokenizer, max_len=128):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len
        # ν¨λ”© ν† ν° ID ν™•λ³΄ (μ—†μΌλ©΄ 0λ²)
        self.pad_token_id = tokenizer.pad_token_id if tokenizer.pad_token_id is not None else 0

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        item = self.data[index]
        text = item['text']
        labels = item['labels']
        words = text.split()
        
        token_list = []
        label_ids = []

        # 1. ν† ν°ν™” λ° λΌλ²¨ λ§¤ν•‘
        for word, label in zip(words, labels):
            word_tokens = self.tokenizer.tokenize(word)
            if not word_tokens: continue
            
            token_list.extend(word_tokens)
            try:
                label_id = LABEL2ID[label]
            except KeyError:
                label_id = LABEL2ID["O"]
                
            label_ids.append(label_id)
            label_ids.extend([-100] * (len(word_tokens) - 1))

        # 2. [ν•µμ‹¬ μμ •] κ°•μ  κΈΈμ΄ λ§μ¶”κΈ° (Truncation & Padding)
        # κΈΈμ΄κ°€ MAX_LENλ³΄λ‹¤ κΈΈλ©΄ μλ¥΄κΈ°
        if len(token_list) > self.max_len - 2: # CLS, SEP μλ¦¬ ν™•λ³΄
            token_list = token_list[:self.max_len - 2]
            label_ids = label_ids[:self.max_len - 2]

        # 3. ID λ³€ν™ λ° νΉμ ν† ν° μ¶”κ°€
        input_ids = self.tokenizer.convert_tokens_to_ids(token_list)
        
        # [CLS] ... [SEP] μ¶”κ°€
        input_ids = [self.tokenizer.cls_token_id] + input_ids + [self.tokenizer.sep_token_id]
        label_ids = [-100] + label_ids + [-100]
        
        # 4. ν¨λ”© μ±„μ°κΈ° (λ¬΄μ΅°κ±΄ MAX_LEN κΈΈμ΄κ°€ λλ„λ΅)
        pad_len = self.max_len - len(input_ids)
        if pad_len > 0:
            input_ids += [self.pad_token_id] * pad_len
            label_ids += [-100] * pad_len
            attention_mask = [1] * (len(input_ids) - pad_len) + [0] * pad_len
        else:
            attention_mask = [1] * self.max_len

        return {
            'input_ids': torch.tensor(input_ids, dtype=torch.long),
            'attention_mask': torch.tensor(attention_mask, dtype=torch.long),
            'labels': torch.tensor(label_ids, dtype=torch.long)
        }

# ====================================================
# 3. ν•™μµ λ° μ¶”λ΅  λ΅μ§
# ====================================================
def run_training_and_inference():
    print(f"π€ [Step 1] λ°μ΄ν„° λ΅λ“: {TRAIN_FILE}")
    if not os.path.exists(TRAIN_FILE):
        print("β ν•™μµ λ°μ΄ν„° νμΌ μ—†μ")
        return

    with open(TRAIN_FILE, 'r', encoding='utf-8') as f:
        raw_data = json.load(f)

    # λ¨λΈ & ν† ν¬λ‚μ΄μ €
    model_name = "monologg/koelectra-base-v3-discriminator"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = ElectraForTokenClassification.from_pretrained(model_name, num_labels=len(LABEL_LIST))
    model.to(DEVICE)

    # λ°μ΄ν„° λ΅λ”
    dataset = NERDataset(raw_data, tokenizer, max_len=MAX_LEN)
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
    optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

    # --- ν•™μµ ---
    print(f"π”¥ [Step 2] ν•™μµ μ‹μ‘ (Epochs: {EPOCHS})")
    model.train()
    
    for epoch in range(EPOCHS):
        total_loss = 0
        for i, batch in enumerate(loader):
            input_ids = batch['input_ids'].to(DEVICE)
            mask = batch['attention_mask'].to(DEVICE)
            labels = batch['labels'].to(DEVICE)

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=mask, labels=labels)
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            
        if (epoch+1) % 5 == 0:
            print(f"   Epoch {epoch+1}/{EPOCHS} Loss: {total_loss/len(loader):.4f}")

    # --- μ¶”λ΅  ---
    print(f"\nπ€ [Step 3] μ¶”λ΅  μ‹μ‘: {os.path.basename(TEST_FILE)}")
    
    # 1. ν…μ¤νΈ νμ‹±
    raw_text = parsing_manager.parse_file(TEST_FILE)
    print(f"   [Input Text]: {raw_text[:80].replace(chr(10), ' ')}...") 

    model.eval()
    
    # 2. ν† ν¬λ‚μ΄μ§•
    inputs = tokenizer(
        raw_text, 
        return_tensors="pt", 
        truncation=True, 
        max_length=512, # μ¶”λ΅ ν•  λ•λ” κΈΈκ² λ΄λ„ λ¨
        padding=True 
    ).to(DEVICE)
    
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.argmax(outputs.logits, dim=2)

    # 3. κ²°κ³Ό ν•΄μ„
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    preds = predictions[0].cpu().numpy()

    extracted_tags = {}
    current_entity = None
    current_word = ""

    for token, pred_idx in zip(tokens, preds):
        if token in ["[CLS]", "[SEP]", "[PAD]"]: continue
        label = ID2LABEL.get(pred_idx, 'O')
        clean_token = token.replace("##", "")

        if label.startswith("B-"):
            if current_entity: extracted_tags.setdefault(current_entity, []).append(current_word)
            current_entity = label.split("-")[1]
            current_word = clean_token
        elif label.startswith("I-") and current_entity == label.split("-")[1]:
            current_word += clean_token
        else:
            if current_entity:
                extracted_tags.setdefault(current_entity, []).append(current_word)
                current_entity = None
                current_word = ""
    if current_entity: extracted_tags.setdefault(current_entity, []).append(current_word)

    # 4. JSON μ¶λ ¥
    final_json = map_to_form(extracted_tags)
    print("\nπ“ [Step 4] μµμΆ… κ²°κ³Ό:")
    print(json.dumps(final_json, indent=4, ensure_ascii=False))

def map_to_form(tags):
    """ 
    [μ—…κ·Έλ μ΄λ“] μ¶”μ¶λ νƒκ·Έ λ¦¬μ¤νΈλ¥Ό κΉ”λ”ν• λ¬Έμμ—΄λ΅ λ³€ν™ν•μ—¬ JSON λ§¤ν•‘ 
    """
    form = {
        "basic_info": {"product_type": "overseas"},
        "location_info": {"city": "", "departure_port": "ICN"},
        "product_info": {"product_name": "", "period": ""},
        "hotels": [{"name": "", "grade": ""}],
        "golf_courses": [{"name": "", "option": ""}],  # κ³¨ν”„μ¥ μ •λ³΄ μ¶”κ°€
        "flight": {"airline": "", "number": ""},
        "price_info": "",
        "details": {"inclusions": [], "exclusions": []}
    }

    # Helper: λ¦¬μ¤νΈλ¥Ό λ¬Έμμ—΄λ΅ μμκ² ν•©μΉλ” ν•¨μ
    # μ: ['λ κ°€μ‹', 'λ¦¬μ΅°νΈ'] -> "λ κ°€μ‹ λ¦¬μ΅°νΈ"
    def get_text(tag_key):
        return " ".join(tags.get(tag_key, []))

    # 1. μ§€μ—­ λ° μ¶λ°μ§€
    form["location_info"]["city"] = get_text("CITY")
    form["location_info"]["departure_port"] = get_text("DEPART_PORT")

    # 2. μƒν’λ… λ° κΈ°κ°„
    # μƒν’λ… νƒκ·Έκ°€ μ—†μΌλ©΄ νΈν…”λ… λ“±μΌλ΅ λ€μ²΄
    prod_name = get_text("PRODUCT_NAME")
    if not prod_name and tags.get("HOTEL_NAME"):
        prod_name = f"{get_text('HOTEL_NAME')} ν¨ν‚¤μ§€"
    form["product_info"]["product_name"] = prod_name
    form["product_info"]["period"] = get_text("PERIOD")

    # 3. νΈν…” μ •λ³΄
    form["hotels"][0]["name"] = get_text("HOTEL_NAME")
    form["hotels"][0]["grade"] = get_text("HOTEL_GRADE")

    # 4. κ³¨ν”„μ¥ μ •λ³΄
    form["golf_courses"][0]["name"] = get_text("GOLF_NAME")
    form["golf_courses"][0]["option"] = get_text("GOLF_OP")

    # 5. ν•­κ³µ μ •λ³΄
    form["flight"]["airline"] = get_text("FLIGHT_NAME")
    form["flight"]["number"] = get_text("FLIGHT_NUM")

    # 6. ν¬ν•¨/λ¶ν¬ν•¨ (λ¦¬μ¤νΈ κ·Έλ€λ΅ μ μ§€)
    form["details"]["inclusions"] = tags.get("INCLUSION", [])
    form["details"]["exclusions"] = tags.get("EXCLUSION", [])

    # 7. κ°€κ²© μ •λ³΄
    form["price_info"] = get_text("PRICE")

    return form

if __name__ == "__main__":
    run_training_and_inference()
```
